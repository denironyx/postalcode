```python
import pandas as pd

# Assuming 'df' is your DataFrame containing the data
# Replace NaN values with None for easier handling
df = df.where(pd.notna(df), None)

# Group by 'Area' and 'Postal Code'
grouped = df.groupby(['Area', 'Postal Code'])
```

```python
# Iterate over each group
for name, group in grouped:
    # Check if any row in the group has valid latitude and longitude
    valid_row = group[(group['Latitude'].notnull()) & (group['Longitude'].notnull())]
    
    if not valid_row.empty:
        # Get the latitude and longitude values from the first valid row
        lat = valid_row.iloc[0]['Latitude']
        lon = valid_row.iloc[0]['Longitude']
        
        # Fill NaN values in the group with latitude and longitude
        group['Latitude'] = group['Latitude'].fillna(lat)
        group['Longitude'] = group['Longitude'].fillna(lon)
        
        # Update the original DataFrame with the filled values
        df.loc[group.index, ['Latitude', 'Longitude']] = group[['Latitude', 'Longitude']]

# Convert back None values to NaN
df = df.where(pd.notna(df), None)
```

```python
df.head()
```

```python
df.to_csv('../data/final_df.csv', index=False)
```
